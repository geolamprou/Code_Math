**Mean Value**

Mean Value is a very bacic function not only in Maths but in every computational Science. 

The method of computing mean value is very simple. We add all the numbers and then we devide then by their number.

If $a_i \in \mathbb{R}, i = 1,2,...,n$, then the type of mean value is presented below:

$$\bar{X}  = \dfrac{a_1 +a_2 +a_3 + ... + a_n}{n}$$

In this context, the code below describes the computation of mean value with 2 ways:
1. Function builted "by hand"
2. Built-in Python function 

```python
# 1. Function builted "by hand" (Example with 3 numnbers)
number_a = float(input("Type the first number: "))
number_b = float(input("Type the second number: "))
number_c = float(input("Type the third number: "))

def mean_of_3(a,b,c):
    if type(a) != str and  type(b) != str and  type(c) != str:
        mean_value = (a+b+c) / 3
        print(f"The mean value of number: {a}, {b}, {c} is equal to {round(mean_value,2)}")
    else:
        print("Check up your input!")

mean_of_3(number_a, number_b, number_c)
```

```python
# 2. Built-in Python Function (using module Statistics)
import statistics

data = [number_a, number_b, number_c]
mean_value_2 = statistics.mean(data)
print(f"The mean value of number: {number_a}, {number_b}, {number_c} is equal to {round(mean_value_2,2)}")
```
