The Equation $a \cdot x = b$.

We are going to solve the equation $a \cdot x = b$. using the programming language Python. 

However, before the code's presentation, let's present the theoritical background of this equation.
For start, let's assume that $a,b \in \mathbb{R}$. We have 3 possible circumstances:

1. If $a=0 \wedge b=0$ then the equation has infinite solutions.
2. If $a=0 \wedge b \neq 0$ then the equation has no sulution.
3. If $a\neq 0$ then the equation has only one solution, the value: $x=\dfrac{b}{a}$.

Based on this, the corresponding code is presented below.

```python
a = float(input("Value of a: "))
b = float(input("Value of b: "))

if a==0.0 and b!=0:
    print("The equation has no solution.")
elif a==0.0 and b==0.0:
    print("The equation has inifinite solutions.")
elif a!=0 and b!=0:
    solution= b / a
    print(f"The equation has only one sulution, the value: {solution} ")
```
