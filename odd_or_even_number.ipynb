Natural Numbers are categorized in to Odd and Even numbers. For example: 4 is an odd number and 7 is an even number.

The folllowing script aims to detect when a number is odd or even.
The method that the code based on, is presented below:

1. Odd number is called a number which when divided by 2 gives a quotient of 0.

2. Even number is called a number which when divided by 2 gives a quotient of 1.

```python
number = int(input("Type a number: "))
quotient = number % 2

if quotient == 0:
    print(f"The number {number} is odd")
else:
    print(f"The number {number} is even")
```
