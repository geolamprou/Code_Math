### Prime Numbers
Prime number is a number that devided perfectly only by itself and by the number 1. 
For example, some of the prime numbers are: 2,3,5,7,11,13,17,19,....

So, in the script below, it will be presented a function that exports if a numbers is prime or not!

```python
def is_prime(num):
    if num <= 5:
        if num == 1 or num ==4:
            print(f"The number {num} is not a prime number!")
        else: 
            print(f"The number {num} is a prime number!")
    else:
        if num % num ==0 and num % 1 ==0 and (num % 2 == 0 or num % 3 == 0 or num % 5 == 0 or 7 == 0):
            print(f"The number {num} is not a prime number!")
        else:
            print(f"The number {num} is a prime number!")
```

To test the function is_prime( ), we are going to check it with 2 numbers: 7 and 14. The correct result is that 7 is a prime number and 14 is not a prime number. 

```python
# Check with the number 7
is_prime(7)
```

```python
# Check with the number 14
is_prime(14)
```
